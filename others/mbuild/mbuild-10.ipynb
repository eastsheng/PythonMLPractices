### PVP二聚体

```python
import mbuild as mb
```

```python
from mbuild.lib.recipes.polymer import Polymer
```

```python
PVP = mb.load('C=CN1CCCC1=O', smiles=True)
```

```python
PVP.visualize()
```

```python
# import numpy as np
import random
chain = Polymer()
delta=0.4
for around in [(1, 0, 0), (0, 1, 0), (0, 0, 1)]:  # rotate around x, y, and z
    # PVP.rotate(random.uniform(0, np.pi), around)
    PVP.rotate(random.uniform(-delta,delta), around)
# PVP.rotate(np.pi/3, [1, 1, 1])
chain.add_monomer(compound=PVP,
                  indices=[8, 10],
                  separation=.15,
                  replace=True,
                  # orientation=[[1,0,0],[1,1,0]],
                 )

# chain.build(n=4, sequence='A')
chain.build(n=4)
chain.energy_minimize()
chain.visualize()
```

```python
chain.visualize()
```
