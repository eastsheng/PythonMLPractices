```python
# 深度学习
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
# !dir "cifar-10-batches-py"
img_data = unpickle('cifar-10-batches-py/data_batch_1')
```
