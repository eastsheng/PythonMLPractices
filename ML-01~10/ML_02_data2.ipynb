```python
# 读取本地鸢尾花数据，并给列名
import pandas as pd
PATH = './'
df = pd.read_csv(PATH+'iris.data',names=['sepal length','sepal width','petal length','petal width', 'class'])
```

```python
df.head()
```

```python
# 按class分组,平均
df.groupby("class").mean()
```

```python
# 按class分组,描述
df.groupby("class").describe()
```
