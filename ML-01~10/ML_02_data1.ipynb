```python
# 散点图
fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(df['petal width'],df['petal length'],color='red')
ax.set_xlabel("Petal Width",fontsize=12)
ax.set_ylabel("Petal Length",fontsize=12)
plt.title('Iris Petal Scatter Diagram',fontsize=14,y=1.01)
```

```python
# 曲线
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df['petal length'],color='green')
ax.set_xlabel("Specimen Number",fontsize=12)
ax.set_ylabel("Petal Length",fontsize=12)
plt.title('Iris Petal Length Curve',fontsize=14,y=1.01)
```

```python
# 条形图：将使用三类鸢尾花中每个特征的平均值绘制一个条形图
bar_width = 0.5
labels = [x for x in df.columns if 'length' in x or 'width' in x]
```

```python
labels
```

```python
ver_y = [df[df['class'] == 'Iris-versicolor'][x].mean() for x in labels]
ver_y
```

```python
vir_y = [df[df['class'] == 'Iris-virginica'][x].mean() for x in labels]
vir_y
```

```python
set_y = [df[df['class'] == 'Iris-setosa'][x].mean() for x in labels]
set_y
```

```python
fig, ax = plt.subplots(figsize=(6,4))
x = np.arange(1,len(labels)+1)
# bottom参数将该序列的y点最小值设置为其下面那个序列的y点最大值
ax.bar(x, vir_y, bar_width, bottom=set_y,color="darkgrey")
ax.bar(x, set_y, bar_width, bottom=ver_y,color="white")
ax.bar(x, ver_y, bar_width, color="blue")
ax.set_xticks(x + (bar_width/20))
ax.set_xticklabels(labels, rotation=-0, fontsize=12);
ax.set_title('Mean Feature Measurement By Class', y=1.01)
ax.legend(['Virginica','Setosa','Versicolor'])
```

```python
## 数据处理
### map方法改变class列
df['class'] = df['class'].map({'Iris-setosa': 'SET', 'Iris-virginica': 'VIR', 'Iris-versicolor': 'VER'})
```

```python
df.head()
```

```python
### apply方法
#### 用apply方法，根据花萼宽创建新列
df['wide petal'] = df['petal width'].apply(lambda v: 1 if v>=1.3 else 0)
df
```

```python
### apply直接作用在dataframe上
#### 增加一个新特征
df['petal area'] = df.apply(lambda r: r['petal length']*r['petal width'],axis=1)
df
```

```python
df.apply('petal area')
```

```python
### applymap 函数
#### 对整个dataframe执行对数函数
df = df.applymap(lambda v: np.log(v) if isinstance(v,float) else v)
df.head()
```
