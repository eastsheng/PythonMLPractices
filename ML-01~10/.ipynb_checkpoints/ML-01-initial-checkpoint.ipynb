## 机器学习
### 获取数据
- 可以来源于csv文件、服务器数据、爬虫数据、图像、视频等
### 检查和探索数据
- 检查不可能地数据；
- 是否有意义；
- 是否缺失或不完整；
### 数据清洗和准备
- 转换为适合于模型使用地格式；
- 过滤、聚集、输入、转化；
### 模型训练
- 选择适当地算法；
- 数据分割为训练、测试和验证集合；
- 避免过拟合等；
### 模型评估
- 根均方误差、欧几里德距离、F1得分；
- 实际值与预估计值的距离量度；
### 部署与应用
- 一旦模型好的模型，就可以应用了。

```python
# 数据可视化
import matplotlib.pyplot as plt
plt.style.use('ggplot') #设置类似R语言中的ggplot库
# 设置插图
%matplotlib inline 
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['petal width'],color='blue')
ax.set_xlabel("Width",fontsize=12)
ax.set_ylabel("Count",fontsize=12)
plt.title('Iris Petal Width',fontsize=14,y=1.01)
```

```python
df.columns
```

```python
fig, ax = plt.subplots(2,2,figsize=(6,4))
fig.tight_layout()
plt.subplots_adjust(wspace =0.25, hspace =0.8)
for i in range(2):
    ax[0][i].hist(df[df.columns[i]],color='blue')
    ax[0][i].set_xlabel(df.columns[i],fontsize=12)
    ax[0][i].set_ylabel("Count",fontsize=12)
    ax[0][i].set_title(df.columns[i],fontsize=14,y=1.01)
for i in range(2):
    ax[1][i].hist(df[df.columns[i+2]],color='blue')
    ax[1][i].set_xlabel(df.columns[i+2],fontsize=12)
    ax[1][i].set_ylabel("Count",fontsize=12)
    ax[1][i].set_title(df.columns[i+2],fontsize=14,y=1.01)
```

```python
# 散点图
fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(df['petal width'],df['petal length'],color='red')
ax.set_xlabel("Petal Width",fontsize=12)
ax.set_ylabel("Petal Length",fontsize=12)
plt.title('Iris Petal Scatter Diagram',fontsize=14,y=1.01)
```

```python
# 曲线
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df['petal length'],color='green')
ax.set_xlabel("Specimen Number",fontsize=12)
ax.set_ylabel("Petal Length",fontsize=12)
plt.title('Iris Petal Length Curve',fontsize=14,y=1.01)
```
