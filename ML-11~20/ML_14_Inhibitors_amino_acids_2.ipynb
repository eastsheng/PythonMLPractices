```python
from docx import Document
SI = "../data/1-s2.0-S0167732222024916-mmc1.docx"
doc = Document(SI)
tables = doc.tables
```

```python
table = tables[0]
```

```python
# save table to cvs
data = open("../data/amino_acids_hydrate_data.csv","w")
for i in range(len(table.rows)):
    for j in range(len(table.rows[0].cells)):
        if j==len(table.rows[0].cells)-1:
            data.write(str(table.cell(i,j).text))
        else:
            data.write(str(table.cell(i,j).text)+",")
    data.write("\n")
data.close()
```

```python
df.index = df["Amino Acid"]
df.drop("Amino Acid",axis=1,inplace=True)
df.drop("No. ",axis=1,inplace=True)
df
```

```python
x = df[['MW', 'pI', 'hI', 'conc ', 'C1%', 'C2%', 'C3%', 'P']]
```

```python
y = df[['T']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=22)
```

```python
# 标准化
from sklearn.preprocessing import StandardScaler
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# # 线性回归
from sklearn.linear_model import LinearRegression
estimator = LinearRegression()

# from sklearn.linear_model import SGDRegressor
# estimator = SGDRegressor(max_iter=1000)
from sklearn.linear_model import Ridge, ElasticNet, Lasso
from sklearn import metrics
import numpy as np
# estimator = Ridge(alpha=1) # 岭回归

# from sklearn.svm import LinearSVC #, SVC, SVR
# estimator = LinearSVC()
estimator.fit(x_train,y_train)
y_predict = estimator.predict(x_test)

score = estimator.score(x_test, y_test)
RS = metrics.r2_score(y_test,y_predict)
MSE = metrics.mean_squared_error(y_test, y_predict)
RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print("准确率为：\n", score)
print("R^2：\n", RS)
print("MSE：\n", MSE)
print("RMSE：\n", RMSE)

a = estimator.coef_
b = estimator.intercept_
```

```python
print("回归系数：",a)
print("截距：",b)
```
