## 深度学习基础
### 线性回归
- 单层神经网络

```python
# 只利用Tensor和GradientTape来实现一个线性回归的训练
%matplotlib inline
import tensorflow as tf
print(tf.__version__)
from matplotlib import pyplot as plt
import random
```

#### 1. 生成简单的人工数据集

```python
# 定义一些变量
num_inputs = 2 # 特征数目
num_examples = 1000 # 样本数目
true_w = [2, -3.4] # 回归模型权重
true_b = 4.2 # 回归模型偏差
# 随机生成服从正态分布的标准差为1、特征数为2的1000个输入样本
features = tf.random.normal((num_examples, num_inputs),stddev = 1)
```

```python
features
```

```python
# 目标值生成
labels = true_w[0] * features[:,0] + true_w[1] * features[:,1] + true_b
# 添加噪声
labels += tf.random.normal(labels.shape,stddev=0.01)
```

```python
features[0],labels[0]
```

```python
# 观察两个输入特征分别与目标之间的关系
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(8,4))
ax1.scatter(features[:,0], labels,1)
ax2.scatter(features[:,1], labels,1)
ax1.set_xlabel("feature1")
ax1.set_ylabel("labels")
ax2.set_xlabel("feature2")
```

#### 2. 不断读取小批量数据样本

```python
# 定义一个函数迭代读取
def data_iter(batch_size, features, labels):
    '''
    batch_size: 小批量的大小；
    features：特征数据集；
    labels：目标数据集
    '''
    num_examples = len(features)
    indices = list(range(num_examples)) # 生成索引
    random.shuffle(indices) # 打乱索引
    for i in range(0, num_examples, batch_size):
        j = indices[i: min(i+batch_size, num_examples)]
        yield tf.gather(features, axis=0, indices=j), tf.gather(labels, axis=0, indices=j)
```

```python
# 读取
batch_size = 10
for X, y in data_iter(batch_size, features, labels):
    print(X, y)
    break
```

#### 3. 初始化模型参数并定义模型

```python
# 将权重初始化成均值为0、标准差为0.01的正态随机数，偏差则初始化成0
w = tf.Variable(tf.random.normal((num_inputs, 1), stddev=0.01))
b = tf.Variable(tf.zeros((1,)))
```

```python
w,b
```

```python
# 定义线性回归模型
def linreg(X, w, b):
    """
    X：输入特征
    w：权重
    b：偏置
    """
    return tf.matmul(X, w) + b # matmul做矩阵乘法，权重与输入相乘加上偏置
```

#### 4. 定义损失函数

```python
# 使用平方损失来定义线性回归的损失函数
# 需要把真实值y变形成预测值y_hat的形状
def squared_loss(y_hat, y):
    return (y_hat - tf.reshape(y, y_hat.shape)) ** 2 /2
```

#### 5. 定义优化算法
- 小批量随机梯度下降算法

```python
def sgd(params, lr, batch_size, grads):
    """Mini-batch stochastic gradient descent.
    lr：学习率
    batch_size：批量大小
    """
    for i, param in enumerate(params):
        param.assign_sub(lr * grads[i] / batch_size)
```

#### 6. 训练模型

```python
lr = 0.03 # 学习率
num_epochs = 3 # 迭代周期个数
net = linreg # 模型
loss = squared_loss # 损失函数

for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        with tf.GradientTape() as t:
            t.watch([w,b])
            l = tf.reduce_sum(loss(net(X, w, b), y))
        grads = t.gradient(l, [w, b])
        sgd([w, b], lr, batch_size, grads)
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, tf.reduce_mean(train_l)))
```
