```python
# 氨基酸作为抑制剂
import pandas as pd
from chembl_webresource_client.new_client import new_client
# 搜索
target = new_client.target
target_query = target.search('amino acids')
targets = pd.DataFrame.from_dict(target_query)
targets.head()
```

```python
type(targets)
```

```python
df = targets[targets.score.notna()]
df
```

```python
df = df[df.tax_id.notna()]
df
```

```python
df.to_csv('../data/inhibitor_amino_acids_data.csv', index=False)
!ls -l ../data
```

```python
df.score.max(),df.score.min()
```

```python
good_amino = []
for i in df.score:
    if float(i)<11:
        good_amino.append("good")
    elif float(i)>=11:
        good_amino.append("not good")
```

```python
selection = ['target_chembl_id', 'species_group_flag',"tax_id", 'score']
df1 = df[selection]
df1
```

```python
df2 = pd.concat([df1,pd.Series(good_amino,name="good_amino")],axis=1)
df2.head(),df2.tail()
```
