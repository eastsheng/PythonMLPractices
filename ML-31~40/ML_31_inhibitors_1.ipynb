```python
import pandas as pd
```

```python
from chembl_webresource_client.new_client import new_client
```

```python
# 搜索抑制剂
target = new_client.target
target_query = target.search('amino acid')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# 保存数据
targets.to_csv("../data/inhibitors_data/amino_acid_950.csv",index=False)
```
