```python
# High-throughput screening
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
# PubChemPy提供了一种Python与PubChem进行交互的方法
import pubchempy as pcp
```

```python
# 通过单体搜索
ph_bpin = pcp.get_compounds('Vinylpyrrolidone', 'name')
ph_bpin
```

```python
ph_bpin = ph_bpin[0]
tmp = Chem.MolFromSmiles(ph_bpin.canonical_smiles)
tmp
```

```python
# pubchem数据库加载
# PubChem数据库包括 3个子数据库：
# 1. PubChem BioAssay 库用于存储生化实验数据，实验数据主要来自高通量筛选实验和科技文献；
# 2. PubChem Compound 库用于存储整理后的化合物化学结构信息；
# 3. PubChem Substance 用于存储机构和个人上传的化合物原始数据。
ph_bpin.canonical_smiles
```

```python
def get_type_compounds(smiles,MaxRecords,searchtype='similarity'):
   
    compounds = pcp.get_compounds(smiles, 'smiles', searchtype=searchtype, MaxRecords=MaxRecords)

    return compounds
```

```python
MaxRecords = 5000
compounds = get_type_compounds(ph_bpin.canonical_smiles,MaxRecords,searchtype="similarity")
compounds
```

```python
mols = []
for i in range(len(compounds)):
    name=compounds[i].iupac_name
    smiles=compounds[i].isomeric_smiles
    formula=compounds[i].molecular_formula
    weight=compounds[i].molecular_weight
    mol = Chem.MolFromSmiles(smiles)
    mols.append(mol)
    print(i,name,smiles,formula,weight)
Draw.MolsToGridImage(mols)
```

```python
MaxRecords = 2000
# searchtype = "similarity"
# searchtype = "superstructure"
searchtype = "substructure"
compounds = get_type_compounds(ph_bpin.canonical_smiles,MaxRecords,searchtype)
number_of_compounds = len(compounds)

mols = []
for i in range(number_of_compounds):
    name=compounds[i].iupac_name
    smiles=compounds[i].isomeric_smiles
    formula=compounds[i].molecular_formula
    weight=compounds[i].molecular_weight
    # print(i,name,smiles,formula,weight)
    if float(weight)<float(ph_bpin.molecular_weight)*10:
        mol = Chem.MolFromSmiles(smiles)
        mols.append(mol)
print(number_of_compounds,len(mols))
Draw.MolsToGridImage(mols)
```

```python
df = pcp.compounds_to_frame(compounds, properties=['iupac_name', 'molecular_formula','isomeric_smiles',"molecular_weight"])
df
```
