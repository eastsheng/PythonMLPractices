```python
# tensorflow2.x 水合物抑制剂线性回归
# 1. 数据读取
import tensorflow as tf
from tensorflow.keras import models,layers,losses,metrics,optimizers
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
# read amino_acids_hydrate_data.csv
df = pd.read_csv("../data/amino_acids_hydrate_data.csv")
df
```

```python
tf.print(df.shape)
df.index = df["Amino Acid"]
df.drop("Amino Acid",axis=1,inplace=True)
df.drop("No. ",axis=1,inplace=True)
tf.print(df.shape)
df
```

```python
SMILES = {"Glycine":"C(C(=O)O)N","Alanine":"C[C@@H](C(=O)O)N","Serine":"C([C@@H](C(=O)O)N)O","Proline":"C1C[C@H](NC1)C(=O)O",
         "Arginine":"C(C[C@@H](C(=O)O)N)CNC(=N)N","Lysine":"C(CCN)C[C@@H](C(=O)O)N","L-valine":"CC(C)[C@@H](C(=O)O)N",
         "L-threonine":"C[C@H]([C@@H](C(=O)O)N)O","L-asparagine":"O=C(N)C[C@H](N)C(=O)O","L-phenylalanine":"c1ccc(cc1)C[C@@H](C(=O)O)N",
         "Valine":"CC(C)[C@@H](C(=O)O)N"}
```

```python
smiles = []
for sm in df.index:
    smiles.append(SMILES[sm])
len(smiles)
df_smiles = pd.Series(smiles,name="Smiles")
df_smiles
```

```python
# notation which contains exact atomic details of molecule
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
import numpy as np
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df_smiles)
df_lipinski
```

```python
df_lipinski.drop("MW",axis = 1, inplace=True)
df_lipinski.index = df.index
df_lipinski
```

```python
df
```

```python
df = pd.concat([df,df_lipinski],axis=1)
tf.print(df.shape)
df
```

```python
x = df[['MW', 'pI', 'hI', 'conc ', 'C1%', 'C2%', 'C3%', 'P','LogP','NumHDonors', 'NumHAcceptors']]
y = df[['T']]
x.shape
```

```python
x
```

```python
x_data = tf.convert_to_tensor(x.values, dtype=tf.float32)
y_data = tf.convert_to_tensor(y.values, dtype=tf.float32)
```

```python
x_data.shape,y_data.shape
```

```python
# 2. 划分数据集
x_train, x_test= tf.split(  
           x_data,
           num_or_size_splits=[160, 50],
           axis=0
       )
y_train, y_test = tf.split(
           y_data,
           [160, 50],
           axis=0
       )
x_train = x_train.numpy()
x_test = x_test.numpy()
x_train.shape
```

```python
# # # 标准化
# from sklearn.preprocessing import StandardScaler
# transfer = StandardScaler()
# x_train = transfer.fit_transform(x_train)
# x_test = transfer.fit_transform(x_test)
```

```python
# 3. 定义模型
tf.keras.backend.clear_session()

model = models.Sequential()
model.add(layers.BatchNormalization())
model.add(layers.Dense(1,activation = tf.nn.relu,input_shape=(11,)))
# print("model.input_shape: ",model.input_shape)
# print("model.output_shape: ",model.output_shape)
# model.summary()
```

```python
# 4. 训练模型
model.compile(optimizer='adam', # SGD Adagrad RMSprop Adadelta Adam Nadam
            loss='mse', # mse mean_squared_error 
            metrics=['mae']) # mae Mean Absolute Error

history = model.fit(
                    x_train,y_train,
                    batch_size = 10,
                    epochs = 200,
                    validation_split=0.1
                   )
# tf.print("w = ",model.layers[0].kernel)
# tf.print("b = ",model.layers[0].bias)
```

```python
# 5. 结果可视化
y_predict = model.predict(x_test)
# tf.print(y_predict)
# w,b = model.variables

# plt.figure(figsize = (12,5))
ax1 = plt.subplot(111)
ax1.scatter(y_test,y_predict, c = "b",label = "samples")
ax1.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],"-r",linewidth = 2.0,label = "model")
ax1.legend()
# ax1.set_xlim(275,287)
# ax1.set_ylim(275,287)
plt.xlabel("Tested Tremperature (K)")
plt.ylabel("Predicted Temperature (K)")
```

```python
plt.plot(history.history['mae'])
plt.plot(history.history['val_mae'])
plt.legend(['training', 'valivation'], loc='best')
```

```python
result = model.evaluate(x_test, y_test)
```
